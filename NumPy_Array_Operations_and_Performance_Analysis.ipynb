```python
#Task 1: Array Creation and Properties using NumPy.
import numpy as np

# 1D array of 100 random integers between 1 and 50
array1 = np.random.randint(1, 51, 100)

# 2D array (5,6) filled with ones and converted to integer type
array2 = np.ones((5, 6)).astype(int)

# 1D array of 20 evenly spaced values between 0 and 100
array3 = np.linspace(0, 100, 20)

# 3x3 identity matrix
array4 = np.eye(3)

# Function to print array properties
def print_properties(arr, name):
    print(f"\n{name}")
    print("Array:\n", arr)
    print("Shape:", arr.shape)
    print("Dimensions (ndim):", arr.ndim)
    print("Total elements (size):", arr.size)
    print("Data type (dtype):", arr.dtype)
    print("Item size (bytes):", arr.itemsize)

# Display properties
print_properties(array1, "1D Random Integer Array")
print_properties(array2, "2D Ones Array (Integer)")
print_properties(array3, "Evenly Spaced Array")
print_properties(array4, "3x3 Identity Matrix")
```

```python
#Task 2: Memory and Speed Comparison (Python List vs NumPy Array).
import numpy as np
import sys
import time

# Create Python list of 10,000 integers
python_list = list(range(1, 10001))

# Create NumPy array with same integers
numpy_array = np.arange(1, 10001)

# Memory consumption
list_memory = sys.getsizeof(python_list)
array_memory = numpy_array.nbytes

print("Memory Consumption:")
print("Python List Memory:", list_memory, "bytes")
print("NumPy Array Memory:", array_memory, "bytes")

# Create second list and array for addition
python_list2 = list(range(1, 10001))
numpy_array2 = np.arange(1, 10001)

# List addition timing
start_time = time.time()

result_list = [python_list[i] + python_list2[i] for i in range(len(python_list))]

list_time = time.time() - start_time

# NumPy addition timing
start_time = time.time()

result_array = numpy_array + numpy_array2

numpy_time = time.time() - start_time

# Display results
print("\nExecution Time:")
print("Python List Addition Time:", list_time, "seconds")
print("NumPy Array Addition Time:", numpy_time, "seconds")

# Speed comparison
speed_factor = list_time / numpy_time
print("\nNumPy is", speed_factor, "times faster than Python lists")
```

```python
#Task 3: Practical Application (Sales Data Simulation).
import numpy as np

# Generate sales data: 30 days × 4 products
sales_data = np.random.randint(100, 1001, size=(30, 4))

print("Sales Data:\n", sales_data)

# 1. Total sales for each product (column-wise sum)
product_total_sales = np.sum(sales_data, axis=0)
print("\nTotal Sales for Each Product:", product_total_sales)

# 2. Average daily sales across all products (row-wise mean)
daily_average_sales = np.mean(sales_data, axis=1)
print("\nAverage Daily Sales:", daily_average_sales)

# 3. Day with maximum total sales
daily_totals = np.sum(sales_data, axis=1)
max_sales_day = np.argmax(daily_totals)

print("\nDay with Maximum Total Sales:", max_sales_day + 1)
print("Maximum Sales Value:", daily_totals[max_sales_day])

# 4. Product with minimum average sales
product_avg_sales = np.mean(sales_data, axis=0)
min_avg_product = np.argmin(product_avg_sales)

print("\nProduct with Minimum Average Sales:", min_avg_product + 1)
print("Average Sales:", product_avg_sales[min_avg_product])

# 5. Convert integers to float32 to optimize memory
sales_float32 = sales_data.astype(np.float32)

print("\nOriginal Data Type:", sales_data.dtype)
print("New Data Type:", sales_float32.dtype)

# Memory comparison per element
original_itemsize = sales_data.itemsize
new_itemsize = sales_float32.itemsize

memory_saved = original_itemsize - new_itemsize

print("\nOriginal Memory per Element:", original_itemsize, "bytes")
print("New Memory per Element:", new_itemsize, "bytes")
print("Memory Saved per Element:", memory_saved, "bytes")
```
